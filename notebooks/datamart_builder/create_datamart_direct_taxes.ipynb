## üßæ Tipos impositivos vigentes en 2025 por territorio

### üá™üá∏ IVA (Impuesto sobre el Valor A√±adido) ‚Äì Pen√≠nsula y Baleares

- **Tipo general**: 21%
- **Tipo reducido**: 10%
- **Tipo superreducido**: 4%
- **Tipo cero**: 0%

Aplicaciones comunes:
- **4%**: Alimentos b√°sicos (pan, leche, huevos, frutas, verduras), libros, peri√≥dicos y medicamentos.
- **10%**: Alimentos no b√°sicos, transporte de viajeros, hosteler√≠a, viviendas de nueva construcci√≥n.
- **21%**: Resto de bienes y servicios no incluidos en los anteriores.

### üáÆüá® IGIC (Impuesto General Indirecto Canario) ‚Äì Canarias

- **Tipo general**: 7%
- **Tipo reducido**: 3%
- **Tipo cero**: 0%
- **Tipo incrementado**: 9,5% y 13,5%
- **Tipo especial**: 20%

Aplicaciones comunes:
- **0%**: Bienes y servicios de primera necesidad (alimentos b√°sicos, medicamentos, libros).
- **3%**: Productos textiles o qu√≠micos, suministros.
- **7%**: Tipo general.
- **9,5% / 13,5%**: Veh√≠culos, bebidas alcoh√≥licas, tabaco, productos de lujo.
- **20%**: Tabaco negro y rubio, suced√°neos del tabaco.

### üá™üá¶ IPSI (Impuesto sobre la Producci√≥n, los Servicios y la Importaci√≥n) ‚Äì Ceuta y Melilla

- **Tipo general**: 4%
- **Tipos reducidos**: 0,5%, 1%, 2%
- **Tipos incrementados**: 8%, 10%

Aplicaciones comunes:
- **0,5%**: Viviendas nuevas, protecci√≥n oficial, transporte de viajeros.
- **1%**: Taxis, cafeter√≠as, bares, consumo el√©ctrico.
- **2%**: Restaurantes de dos tenedores o m√°s.
- **4%**: Reformas, mejoras, inmuebles.
- **8% / 10%**: Servicios electr√≥nicos, construcci√≥n.

## üß† Justificaci√≥n de la asignaci√≥n de tipos impositivos a las categor√≠as de gasto

Para asignar los tipos impositivos (IVA, IGIC e IPSI) a las 45 categor√≠as de gasto del datamart, se aplic√≥ una metodolog√≠a basada en:

### 1. Correspondencia conceptual
Se asign√≥ el tipo m√°s coherente con el contenido de cada categor√≠a.  
Ejemplo: `01.1 PRODUCTOS ALIMENTICIOS` ‚Üí IVA 4% por tratarse de alimentos b√°sicos.

- **IGIC**: 0% (alimentos de primera necesidad est√°n exentos).
- **IPSI**: 0,5% (tipo simb√≥lico para alimentos en Ceuta y Melilla).

### 2. Legislaci√≥n oficial consultada
- **IVA**: Ley 37/1992 y sus actualizaciones (BOE).
- **IGIC**: Decreto Legislativo 1/1994 y portal oficial de la Agencia Tributaria Canaria.
- **IPSI**: Ordenanzas fiscales de Ceuta y Melilla (a√±os 2021‚Äì2024).

### 3. Asignaci√≥n funcional
Cuando no exist√≠a equivalencia directa (e.g. `remesas`, `protecci√≥n social`), se asign√≥ el **0%** por su car√°cter **exento o no sujeto**.

Servicios sociales, educativos o sanitarios se consideraron de **inter√©s general** y se les aplic√≥ un tipo reducido o nulo.

### 4. Actualizaci√≥n con valores de 2025
Los tipos asignados reflejan la situaci√≥n normativa vigente en 2025.  
En ausencia de cambios significativos en la legislaci√≥n, se han mantenido los valores de 2023‚Äì2024.

### 5. Adaptaci√≥n de c√≥digos
Los c√≥digos como `"01.1"` fueron convertidos a `"011"` para facilitar su uso como claves en bases de datos y modelos.
